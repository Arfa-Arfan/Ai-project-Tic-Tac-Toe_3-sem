```python
import math  # for minimum  and maximum values
import time  #for ai text dely
#board setup
board = ["-" , "-" , "-" ,
         "-" , "-" , "-" ,
         "-" , "-" , "-" ]
#global variables
CurrentPlayer = "X"
GameRuning = True
Winner = None
MAX_DEPTH = 5

#display board
def printBoard(board):

    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--+")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--+")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print("--+---+--+")


#Take input from user 
def read_int(prompt , lo=None , hi=None):
    # read an integer from user; keep asking until valid and within bounds
    while True:
        try:
            val = int(input(prompt))
            if lo is not None and val < lo:
                print(f"Value must be >= {lo}"); continue
            if hi is not None and val > hi:
                print(f"Value must be <= {hi}"); continue
            return val
        except ValueError:
            # tell user the input was not an integer
            print("Please enter an integer")

def  PlayerInput(board):
    inp = read_int("Select a spot 1-9" , 1 , 9)
    if board[inp-1] == "-":
        board[inp-1] = CurrentPlayer
        return
    else:
        print("Oops player is already at that spot.")
        PlayerInput(board)

#check win or tie
# def checkhorizontal(board):
#     global Winner
#     if board[0] == board[1] == board[2] and board[0] != "-":
#         Winner = board[0]
#         return True
#     elif board[3] == board[4] == board[5] and board[3] != "-":
#         Winner = board[3]
#         return True
#     elif board[6] == board[7] == board[8] and board[6] != "-":
#         Winner = board[6]
#         return True
#     return False
# def checkVertical(board):
#     global Winner
#     if board[0] == board[3] == board[6] and  board[0] != "-":
#         Winner = board[0]
#         return True
#     elif board[1] == board[4] == board[7] and  board[1] != "-":
#         Winner = board[1]
#         return True
#     elif board[2] == board[5] == board[8] and  board[2] != "-":
#         Winner = board[2]
#         return True
#     return False
# def checkdiagnol(board):
#     global Winner
#     if board[0] == board[4] == board[8] and board[0] != "-":
#         Winner = board[0]
#         return True
#     elif board[2] == board[4] == board[6] and board[2] != "-":
#         Winner = board[0]
#         return True
#     return False
#check all win conditions
def checkwin(board):
    global Winner
    #check rows
    for i  in range(0 , 9 , 3):
        if board[i] == board[i+1] == board[i+2] and board[i] != "-":
            Winner = board[i]
            return True
        #check columns 
    for i in range(3):
        if board[i] == board[i+3] == board[i+6] and board[i] != "-":
            Winner = board[i]
            return True
    #check diagonals
    if board[0] == board[4] == board[8] and board[0] != "-":
        Winner = board[0]
        return True
    elif board[2] == board[4] == board[6] and board[2] != "-":
        Winner = board[2]
        return True
    return False
def checktie(board):
    if "-" not in board:
        return True
    return False
def switchPlayer():
    global CurrentPlayer
    if CurrentPlayer =="X":
        CurrentPlayer ="O"
    else:
        CurrentPlayer ="X"

#check win/tie
def checkGameOver():
    global GameRuning
    if checkwin(board):
        print(f"The winner is {Winner}")
        GameRuning  = False
        return True
    elif checktie(board):
        print("It is a tie!")
        GameRuning = False
        return True
    return False
def evaluateBoard(board):
    if checkwin(board):
        return 1 if Winner == "O" else -1
    elif checktie(board):
        return 0

#minmax algorithm
def minimax(board , depth , isMaximizing):
    score = evaluateBoard(board)
    if score == 1 or score == -1 or score == 0 or depth > MAX_DEPTH:
        return score if score is not None else 0
    
    # maximizing part
    if isMaximizing:
        bestScore = -math.inf
        for i in range(9):
            if board[i] == "-":
                board[i] = "O"
                bestScore = max(bestScore , minimax(board , depth + 1 , False))
                board[i] = "-"  
        return bestScore
    # minimizing part
    else:
        bestScore = math.inf
        for i in range(9):
            if board[i] == "-":
                board[i] = "X"
                bestScore = min(bestScore , minimax(board , depth + 1 , True))
                board[i] = "-"
        return bestScore
    
def findBestMove(board):
        bestscore = -math.inf
        bestMove = 0
        for i in range(9):
            if board[i] == "-":
                board[i] = "O"
                score = minimax(board , 0 , False)
                board[i] = "-"
                if score > bestscore:
                    bestscore = score
                    bestMove = i
        return bestMove
def ai_text(text ,dely= 0.05):
    for char in text:
        print(char , end='' , flush=True)
        time.sleep(dely)

#game loop
while True:
    printBoard(board)
    if CurrentPlayer =="X":
        PlayerInput(board)
    else:
        ai_text("Thinking...\n",0.06)
        move = findBestMove(board)
        board[move] = "O"
        printBoard(board)
        ai_text("<---------------------->\n" , 0.01)
    if checkGameOver():
        break
    switchPlayer()
```
